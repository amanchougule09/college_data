```python
n=int(input("Enter number"))
def prime(n):
    if n<2:
        print("Number is not prime")
    flag=False
    for i in range(2,n-1):
        if n%i==0:
            flag=True
            break;
    if flag:
        print("Not Prime")
    else:
        print("Prime")
prime(n)
```
