```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hank', 'Ivy', 'Jack'],
    'Experience': [1.1, 1.5, 3.2, 4.5, 5.9, 7.1, 8.2, 9.0, 10.2, 10.5],
    'Salary': [39343, 46205, 54445, 61111, 81363, 93940, 113812, 105582, 122391, 121872]
}

df = pd.DataFrame(data)

# Features and target
X = df[['Experience']]   # must be 2D
y = df['Salary']

# Model
model = LinearRegression()
model.fit(X, y)
model.coef_
model.intercept_

# Predictions
experience_year = [[2], [4], [6]]
predict = model.predict(experience_year)
print("Predicted salaries:", predict)

# Plot
plt.scatter(df['Experience'], df['Salary'], color='blue')
plt.plot(df['Experience'], model.predict(X), color='red')  # regression line
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
